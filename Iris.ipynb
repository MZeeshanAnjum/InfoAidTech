```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("/content/drive/MyDrive/Internship/IRIS.csv")
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
l.fit_transform(data["species"])
```

```python
X=data.drop("species",axis='columns')
y=data["species"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species",data=data)
```

```python

# Create histograms for each feature
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
sns.histplot(data=data, x="sepal_length", kde=True)
plt.title("Sepal Length Histogram")

plt.subplot(2, 2, 2)
sns.histplot(data=data, x="sepal_width", kde=True)
plt.title("Sepal Width Histogram")

plt.subplot(2, 2, 3)
sns.histplot(data=data, x="petal_length", kde=True)
plt.title("Petal Length Histogram")

plt.subplot(2, 2, 4)
sns.histplot(data=data, x="petal_width", kde=True)
plt.title("Petal Width Histogram")

plt.tight_layout()
plt.show()
```

```python
n=len(np.unique(y))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n)
model.fit(X_train,y_train)
```

```python
ypred=model.predict(X_test)
model.score(X_test,y_test)
```
